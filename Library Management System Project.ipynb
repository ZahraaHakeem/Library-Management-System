```python
def printMenu():
    print("\nLibrary Management System\n========================== ")
    print(" 1. Print books Info \n 2. Search a book \n 3. Add new book \n 4. Remove a book \n 5. Borrow a book \n 6. Return a book \n 7. Exit")
    print("==========================")
    
def main():
    allBooks = []
    borrowData = []
    booksFile = open("booksInfoFinalProject.txt","r")
    for line in booksFile:
        line = line.strip()
        wordLine = line.split(",")
        allBooks.append(wordLine)
        
    borrowFile = open("borrowedInfoFinalProject.txt","r")
    for line in borrowFile:
        line = line.strip()
        wordLine = line.split(",")
        borrowData.append(wordLine)

    userChoice = int(input("Enter your choice: ")) 
    print("\n")
    checkUserChoice(userChoice,allBooks,borrowData)
    
def checkUserChoice(userChoice,allBooks,borrowData):
    if userChoice ==1:
        PrintBooksInfo(allBooks)
        
    elif userChoice ==2:
        userChoiceSearchBook = input("Enter (t) to search by title or (a) to search by author name: ")
        if userChoiceSearchBook == "t":
            titleSearch = input("Enter the title: ")
            SearchBookByTitle(allBooks,titleSearch)
            
        elif userChoiceSearchBook == "a":
            authorSearch = input("Enter the author: ")
            SearchBookByAuthor(allBooks,authorSearch)       
      
    elif userChoice ==3:   
        AddNewBook(allBooks)
        
    elif userChoice ==4:   
         RemoveBook(allBooks)
       
    elif userChoice ==5:   
        BorrowBook(borrowData,allBooks)
        
    elif userChoice==6:
        ReturnBook(borrowData,allBooks)
        
    elif userChoice==7:
        UpdtaeFiles(allBooks,borrowData)
        print("The all Files Updated")
   
    while userChoice!=7 :
        printMenu()
        userChoice = int(input("Enter your choice: ")) 
        print("\n")
        checkUserChoice(userChoice,allBooks,borrowData)
        
printMenu()
main()
```

```python
def PrintBooksInfo(allBooks):
    print("Total", len(allBooks) ,  "Books")
    print("-"*50)
    for book in allBooks:
        authorsList = book[2].split(":")
        print("serial#: " , book[0])
        print("title: " , book[1])
        print("number of authors: " , len(authorsList))
        print("price: " ,book[3] )
        print("total copies: " ,int(book[4]) + int(book[5]) )
        print("-"*50)
```

```python
def SearchBookByTitle(allBooks , titleSearch):
    foundResult = False
    print("-"*20)
    foundedBooks = []

    for book in allBooks:
            if titleSearch.upper() in book[1].upper():
                foundResult = True
                foundedBooks.append(book)
  
    if not foundResult:
        print("No matched record found")  
    else:
        print("\nMatched records: ")
        for book in foundedBooks:
            print("serial#: " , book[0])
            print("title: " , book[1])
            print("authors:" )
            authors=book[2].split(":")
            for author in authors:
                print("   -",author)
            print("price: " ,book[3] )
            print("copies in library: " ,int(book[4])  )
            print("borrowed copies: " , int(book[5]) )
            print("-"*50)
```

```python
def SearchBookByAuthor(allBooks, authorSearch):
    foundResult = False
    foundedBooks = []
    for book in allBooks:
        if authorSearch.upper() in book[2].upper():
            foundResult = True
            foundedBooks.append(book)
    if not foundResult:
        print("No matched record found")  
    else:
        print("\nMatched records: ")
        for  book in foundedBooks:
            print("serial#: " , book[0])
            print("title: " , book[1])
            print("authors:" )
            authors=book[2].split(":")
            for author in authors:
                print("   -",author)
            print("price: " ,book[3] )
            print("copies in library: " ,int(book[4])  )
            print("borrowed copies: " , int(book[5]) )
            print("-"*50)
```

```python
def AddNewBook(allBooks): 
    problems= [] 
    problems.clear()
    authorsCount=2
    serial= input("Enter serial number: ")
    title= input("Enter book title: ")
    author=  input("Enter name of author 1: ")
    moreAuthors= input("Enter name of author 2 (Enter q to finish): ")
    while moreAuthors != "q":
        authorsCount+=1
        author = author + ":" + moreAuthors
        moreAuthors= input("Enter name of author %2s (Enter q to finish): " %(authorsCount))
    price= input("Enter book price: ")
    copies= input("Enter number of book copies: ")
    
    serialResult = checkSerialNumber(allBooks,serial) 
    titleResult= checkTitle(title)
    authorsResult = checkAuthors(author)
    priceResult = checkPrice(price)
    copiesResult = checkCopies(copies)
    
    
    if serialResult and titleResult and authorsResult and priceResult and  copiesResult :
        bookData = [serial,title , author,price,copies , 0]
        allBooks.append(bookData)
        print("\n New book was added successfully")
        
    if len(problems)>0:
        print()
        for problem in problems:
            print("Error: "  ,problem)
        
 
    
def checkSerialNumber(allBooks,serial):
    result = True
    if len(serial)!=5:
        problems.append("Serial number must be 5 digits")
        result = False
    else:
        for book in allBooks:
            if serial in book[0]:
                problems.append("Serial number is already used")
                result = False
    return result 


def checkTitle(title):
    result = True
    if title.strip() =="":
        problems.append("Title can't be empty")
        result = False
    return result 

def  checkAuthors(author):
    result = True
    if author.strip()=="":
        problems.append("names of all authors should not be empty ")
        result = False
    return result    
        
def checkPrice(price):
    result = True
    try :
        priceFloat = float(price)
        if priceFloat <=0:
            problems.append("price should be a positive number")
            result = False
    except Exception:
        problems.append("price should be a valid float number ") 
        result = False
    return result  
    
def checkCopies(copies):
    result = True
    try :
        copiesInt= int(copies)
        if copiesInt <=0:
            problems.append("copies should be a positive number") 
            result = False
    except Exception:
        problems.append("number of copies should be a valid int number") 
        result = False
        
    return result
  
```

```python
def RemoveBook(allBooks):
    foundSerialNumber= False
    bookData =[]
    sureToDelete = "y"
    serialToRemove = input("Enter book serial number to remove: ")
    for book in allBooks:
        if serialToRemove== book[0] and foundSerialNumber==False:
            foundSerialNumber = True
            bookData = book

    if  foundSerialNumber == False :
        print("Error: no matched serial number")
            
    else:
            print("Matched records:")
            print("serial#: " , bookData[0])
            print("title: " , bookData[1])
            print("authors:" )
            authors=bookData[2].split(":")
            for author in authors:
                print("   -",author)
            print("price: " ,bookData[3] )
            print("copies in library: " ,int(bookData[4])  )
            print("borrowed copies: " , int(bookData[5]) )
            print("-"*50)
            sureToDelete = input("Deleting the book .. Are you sure (y/n):")

            if sureToDelete.lower()=="n":
                print("operation is cancelled")

            elif  sureToDelete.lower()=="y":
                if int(bookData[5])>0:
                    print("Book cannot be removed: borrowed copies must be 0")
                else:
                    allBooks.remove(bookData)
                    print("Book was removed successfully")

            
           
           
               
```

```python
def BorrowBook(borrowData,allBooks):
    seraialToBorrow = input("Enter book serial number to borrow: ")   
    userID = input("Enter user id who borrowed the book: ")
    sameUserSameBorrowd = False
    sameUserBorroweThreeTimes = False
    foundSerialNumber = False
    serialLengthGood = True

    if len(seraialToBorrow)==5: 
        for book in allBooks:
            if seraialToBorrow== book[0]:
                foundSerialNumber = True
                
        if  foundSerialNumber == False :
            print("Error: no matched serial number")
            
        else :
            for borrow in borrowData:
                if seraialToBorrow== borrow[0] and userID== borrow[1]:
                    sameUserSameBorrowd =True
        
        if sameUserSameBorrowd ==True:
            print("Error: user already borrowed a copy of the book")
            
        elif sameUserSameBorrowd == False and foundSerialNumber==True   :
            count = 0
            for borrow in borrowData:
                if userID == borrow[1]:
                    count+=1
                    if count>2:
                        sameUserBorroweThreeTimes = True
            if sameUserBorroweThreeTimes ==False:
                updateTheCopiesNumber(seraialToBorrow,userID,allBooks,borrowData)
                
            else :
                print("user cannot borrow more than 3 books")

    else:
        print("Error: serial number should be 5 digits")
    
def updateTheCopiesNumber(seraialToBorrow,userID,allBooks,borrowData):
    copiesAvalible = 0
    copiesInBorrowed = 0
    for book in allBooks:
            if seraialToBorrow== book[0]:
                copiesAvalible = int(book[4])
                copiesInBorrowed = int(book[5])
                break
    if copiesAvalible>0:
        book[4]= copiesAvalible - 1
        book[5]=  copiesInBorrowed + 1
        newBorrrow = [seraialToBorrow,userID]
        borrowData.append(newBorrrow)
        print("\n Book was borrowed successfully")
    else:
        print("no available copies in the library")
    
```

```python
def ReturnBook(borrowData,allBooks):
    seraialToReturn = input("Enter book serial number to return: ")   
    userID = input("Enter user id: ")
    borrowRemoved=False
    for borrowRecord in borrowData:   
        if seraialToReturn == borrowRecord[0] and userID == borrowRecord[1] :
            borrowData.remove(borrowRecord)
            borrowRemoved = True
    if borrowRemoved ==True:
        for book in allBooks:
            book[4]= int(book[4]) + 1
            book[5]=  int(book[5]) - 1
        print("Book was returned successfully")
    else:
        print("no matched record found in borrowedInfo.txt")
```

```python
def UpdtaeFiles(allBooks,borrowData):
    bookFile = open("booksInfoFinalProject.txt","w")
    borrowFile = open("borrowedInfoFinalProject.txt","w")
    for book in allBooks:
        for bookDataIndex in range(0,len(book)):
            if bookDataIndex == len(book)-1:
                bookFile.write(str(book[bookDataIndex]))
            else :
                bookFile.write(str(book[bookDataIndex])+",")
        bookFile.write("\n")    
    bookFile.close()
    
    for borrow in borrowData:
        for borrowDataIndex in range(0,len(borrow)):
            if borrowDataIndex == len(borrow)-1:
                borrowFile.write(str(borrow[borrowDataIndex]))
            else :
                borrowFile.write(str(borrow[borrowDataIndex])+",")
        borrowFile.write("\n")    
    borrowFile.close()
```
